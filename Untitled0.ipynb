```python
import json

# Load the JSON data from the file
input_file_path = "crpc_compoundable_offences.json"
output_file_path = "cleaned_output.json"

with open(input_file_path, 'r') as file:
    data = json.load(file)

# Cleaning function to remove \n and \t from the 'compoundable by' field
def clean_compoundable_by_field(data):
    for entry in data:
        if "compoundable by" in entry:
            # Replace \n and \t with a space and strip leading/trailing spaces
            entry['compoundable by'] = entry['compoundable by'].replace('\n', ' ').replace('\t', ' ').strip()
            entry['offence'] = entry['offence'].replace('\n', ' ').replace('\t', ' ').strip()
    return data

# Clean the data
cleaned_data = clean_compoundable_by_field(data)

# Save the cleaned data back to a new JSON file
with open(output_file_path, 'w') as file:
    json.dump(cleaned_data, file, indent=4)

print(f"Data successfully cleaned and saved to {output_file_path}")
```

```python
import pdfplumber
import json

# Define a function to extract table data and convert it to JSON
def extract_table_to_json(pdf_path, page_numbers=[10, 11, 12, 13, 14, 15], output_json="ipcFines.json"):
    data = []

    with pdfplumber.open(pdf_path) as pdf:
        for page_number in page_numbers:
            # Select the page
            page = pdf.pages[page_number - 1]  # Pages are zero-indexed in pdfplumber
            print(f"Extracting table from page {page_number}")

            # Extract the table from the page
            table = page.extract_table()

            # Print the extracted table for debugging
            print(table)

            if table:  # Check if the table is extracted successfully
                column_names = ["Section Number", "Provision/Offense", "Punishment"]

                # Loop through the table rows and create a dictionary for each row
                for row in table[1:]:  # Skip the header row
                    entry = {column_names[i]: row[i] for i in range(len(column_names))}
                    data.append(entry)

    # Convert the list of dictionaries to JSON format
    if data:  # Only write to JSON if data is not empty
        json_data = json.dumps(data, indent=4)
        with open(output_json, "w") as json_file:
            json_file.write(json_data)
        print(f"Data successfully extracted and saved to {output_json}")
    else:
        print("No data extracted. Please check the table extraction process.")

# Example usage
pdf_path = "FinesIPC.pdf"  # Replace with the actual path to your PDF file
extract_table_to_json(pdf_path)

```

```python
import PyPDF2
import json
import re

# Path to the PDF file
pdf_path = "FinesIPC.pdf"

# Open the PDF file
with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    extracted_text = ""

    # Loop through pages 10-15 (pages are 0-indexed, so 9-14)
    for page_num in range(9, 15):
        page = reader.pages[page_num]
        extracted_text += page.extract_text()  # Extract text from the page

# Process the extracted text and convert it into structured data
# We will use regular expressions to identify the section number, provision/offense, and punishment

# Define a regex pattern to match rows in the format (adjust as per your actual format)
pattern = re.compile(r"(\d+)\s+([A-Za-z\s,]+)\s+May extend to Rs\.\s*(\d+)(.*)")

# List to hold structured data
table_data = []

# Split the extracted text by lines and match each line with the regex pattern
lines = extracted_text.split("\n")
for line in lines:
    match = pattern.match(line)
    if match:
        section_number = match.group(1)
        provision_offense = match.group(2).strip()
        punishment = f"May extend to Rs. {match.group(3).strip()}"

        # Add the matched data to the table_data list
        table_data.append({
            "Section number": section_number,
            "Provision/offense": provision_offense,
            "Punishment": punishment
        })

# Convert the table data to JSON format
json_data = json.dumps(table_data, indent=4)

# Save the output to a JSON file
with open("output.json", "w") as json_file:
    json_file.write(json_data)

# Optionally, print the JSON data
print(json_data)
```

```python
import PyPDF2
import json
import re

# Path to the PDF file
pdf_path = "FinesIPC.pdf"

# Open the PDF file
with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    extracted_text = ""

    # Loop through pages 10-15 (pages are 0-indexed, so 9-14)
    for page_num in range(9, 15):
        page = reader.pages[page_num]
        extracted_text += page.extract_text()  # Extract text from the page

# Print extracted text to understand how it looks
# Uncomment this line to inspect the raw text output
# print(extracted_text)

# Enhanced regex pattern to match more variations
pattern = re.compile(r"(\d+)\s+([A-Za-z\s,]+)\s+May extend to Rs\.\s*(\d+)(.*)")

# List to hold structured data
table_data = []

# Split the extracted text by lines
lines = extracted_text.split("\n")

for line in lines:
    # Remove extra spaces from the line
    line = line.strip()

    # Match using the updated regular expression
    match = pattern.match(line)

    if match:
        section_number = match.group(1)
        provision_offense = match.group(2).strip()
        punishment = f"May extend to Rs. {match.group(3).strip()}"

        # Add the matched data to the table_data list
        table_data.append({
            "Section number": section_number,
            "Provision/offense": provision_offense,
            "Punishment": punishment
        })
    else:
        # Print unmatched lines for debugging
        # This will help understand what is missed by the regex
        print(f"Unmatched line: {line}")

# Convert the table data to JSON format
json_data = json.dumps(table_data, indent=4)

# Save the output to a JSON file
with open("output.json", "w") as json_file:
    json_file.write(json_data)

# Optionally, print the JSON data
print(json_data)
```
